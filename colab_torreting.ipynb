# Download

```python
from google.colab import files
import shutil
import os

# Path of the file or folder you want to download
path_to_download = "/content/torrent_download"  # example folder with downloaded files

# If it's a folder, compress it first
if os.path.isdir(path_to_download):
    zip_name = "/content/download.zip"
    shutil.make_archive("/content/download", 'zip', path_to_download)
    files.download(zip_name)
else:
    # If it's a single file, download directly
    files.download(path_to_download)
```

# Cleanup Files

```python
import shutil
import os

# Folder where torrent files are downloaded
download_folder = "/content/torrent_download"

# Remove the torrent download folder if it exists
if os.path.exists(download_folder):
    shutil.rmtree(download_folder)
    print(f"Removed folder: {download_folder}")
else:
    print(f"Folder not found: {download_folder}")

# Remove any .torrent files in the root Colab directory (if any)
for file in os.listdir("/content"):
    if file.endswith(".torrent"):
        os.remove(os.path.join("/content", file))
        print(f"Removed file: {file}")

# Remove download.zip if it exists
zip_path = "/content/download.zip"
if os.path.exists(zip_path):
    os.remove(zip_path)
    print("Removed file: download.zip")

print("Cleanup complete!")
```
